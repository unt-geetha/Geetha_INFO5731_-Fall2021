Question 3 (10 points): Write python code to collect 1000 articles from Google Scholar (https://scholar.google.com/), Microsoft Academic (https://academic.microsoft.com/home), or CiteSeerX (https://citeseerx.ist.psu.edu/index), or Semantic Scholar (https://www.semanticscholar.org/), or ACM Digital Libraries (https://dl.acm.org/) with the keyword "information retrieval". The articles should be published in the last 10 years (2013-2023).

The following information of the article needs to be collected:

(1) Title

(2) Venue/journal/conference being published

(3) Year

(4) Authors

(5) Abstract

```python
# Import necessary libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Define a function to collect article data from Google Scholar
def collect_articles(query, num_articles_to_collect):
    # Set the base URL for Google Scholar
    base_url = "https://scholar.google.com/scholar"
    collected_articles = []

    # Continue collecting articles until the desired number is reached
    while len(collected_articles) < num_articles_to_collect:
        # Define query parameters
        query_params = {
            "q": query,        # Query string
            "hl": "en",        # Language preference (English)
            "start": len(collected_articles),  # Starting index for results
        }

        # Send an HTTP GET request to Google Scholar
        response = requests.get(base_url, params=query_params)
        soup = BeautifulSoup(response.text, "html.parser")

        # Find article entries in the search results
        article_entries = soup.find_all("div", class_="gs_ri")

        for entry in article_entries:
            article_info = {}

            # Extract the article title
            title_element = entry.find("h3", class_="gs_rt")
            article_info["Title"] = title_element.text if title_element else "N/A"

            # Extract the venue or journal
            venue_element = entry.find("div", class_="gs_a")
            article_info["Venue"] = venue_element.text if venue_element else "N/A"

            # Extract the publication year
            year_info = entry.find("div", class_="gs_a").text.split(" - ")[-1]
            article_info["Year"] = year_info if year_info else "N/A"

            # Extract the authors
            authors_info = entry.find("div", class_="gs_a").text.split(" - ")[0]
            article_info["Authors"] = authors_info if authors_info else "N/A"

            # Extract the abstract (if available)
            abstract_element = entry.find("div", class_="gs_rs")
            article_info["Abstract"] = abstract_element.text if abstract_element else "N/A"

            # Add the article data to the list
            collected_articles.append(article_info)

    return collected_articles

# Specify the query and number of articles to collect
search_query = "Data Science"
num_articles_to_collect = 1000

# Collect articles using the defined function
articles_collection = collect_articles(search_query, num_articles_to_collect)

# Print the collected article data
for index, article_data in enumerate(articles_collection, start=1):
    print(f"Article {index}:")
    print(f"Title: {article_data['Title']}")
    print(f"Venue/Journal: {article_data['Venue']}")
    print(f"Year: {article_data['Year']}")
    print(f"Authors: {article_data['Authors']}")
    print(f"Abstract: {article_data['Abstract']}")
    print("\n")







```
