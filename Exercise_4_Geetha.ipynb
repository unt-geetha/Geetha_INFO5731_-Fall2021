## (1) (10 points) Generate K topics by using LDA, the number of topics K should be decided by the coherence score, then summarize what are the topics. You may refer the code here:

> Indented block



https://www.machinelearningplus.com/nlp/topic-modeling-gensim-python/

```python
# Write your code here

# Defining the text corpus
geetha_texts = [
    "Geetha is a data scientist who specializes in natural language processing and machine learning.",
    "In her free time, Geetha enjoys reading books on artificial intelligence and exploring new technologies.",
    "Geetha is dedicated to leveraging technology to solve real-world challenges and make a positive impact.",
    "She is currently engaged in a project involving the development of a chatbot for advanced NLP-based customer support."
    "Geetha has a strong background in computer science and has worked on various data analytics projects.",
    "In addition to her technical skills, Geetha is an avid problem solver and enjoys tackling complex challenges.",
    "She is known for her dedication and innovative approach to problem-solving.",
]

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return tokens

processed_geetha_texts = [preprocess_text(text) for text in geetha_texts]

dictionary = corpora.Dictionary(processed_geetha_texts)
corpus = [dictionary.doc2bow(text) for text in processed_geetha_texts]

coherence_scores = []
start_topics, end_topics = 2, 10

for num_topics in range(start_topics, end_topics + 1):
    lda_model = gensim.models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=5)
    coherence_model = CoherenceModel(model=lda_model, texts=processed_geetha_texts, dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model.get_coherence()
    coherence_scores.append(coherence_score)

num_topics_range = range(start_topics, end_topics + 1)
plt.plot(num_topics_range, coherence_scores)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.title("Optimal Number of Topics")
plt.show()

optimal_num_topics = start_topics + coherence_scores.index(max(coherence_scores))

lda_model = gensim.models.LdaModel(corpus, num_topics=optimal_num_topics, id2word=dictionary, passes=15)

topics = lda_model.print_topics()
for topic in topics:
    print(topic)
```

## (2) (10 points) Generate K topics by using LSA, the number of topics K should be decided by the coherence score, then summarize what are the topics. You may refer the code here:

https://www.datacamp.com/community/tutorials/discovering-hidden-topics-python

```python
# Your Code here

# Defining the text corpus

geetha_texts = [
    "Geetha is a data scientist who specializes in natural language processing and machine learning.",
    "In her free time, Geetha enjoys reading books on artificial intelligence and exploring new technologies.",
    "Geetha is dedicated to leveraging technology to solve real-world challenges and make a positive impact.",
    "She is currently engaged in a project involving the development of a chatbot for advanced NLP-based customer support."
    "Geetha has a strong background in computer science and has worked on various data analytics projects.",
    "In addition to her technical skills, Geetha is an avid problem solver and enjoys tackling complex challenges.",
    "She is known for her dedication and innovative approach to problem-solving.",
]

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return tokens

processed_geetha_texts = [preprocess_text(text) for text in geetha_texts]

dictionary = corpora.Dictionary(processed_geetha_texts)

corpus = [dictionary.doc2bow(text) for text in processed_geetha_texts]

def compute_coherence_values_lsa(dictionary, corpus, texts, limit, start=2, step=1):
    coherence_values = []
    for num_topics in range(start, limit, step):
        lsa_model = LsiModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)
        coherence_model = CoherenceModel(model=lsa_model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_score = coherence_model.get_coherence()
        coherence_values.append((num_topics, coherence_score))
    return coherence_values

limit_topics = 20
start_topics = 2
step_topics = 1

coherence_values_lsa = compute_coherence_values_lsa(dictionary=dictionary, corpus=corpus, texts=processed_geetha_texts, start=start_topics, limit=limit_topics, step=step_topics)

best_num_topics_lsa = max(coherence_values_lsa, key=lambda x: x[1])[0]

lsa_model = LsiModel(corpus=corpus, id2word=dictionary, num_topics=best_num_topics_lsa)

print("Optimal number of (LSA) topics :", best_num_topics_lsa)

topics_lsa = lsa_model.print_topics()
for topic in topics_lsa:
    print(topic)

# Visualizing coherence scores
topics, coherence_scores = zip(*coherence_values_lsa)
plt.plot(topics, coherence_scores)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score (LSA)")
plt.title("Coherence Score vs. Number of Topics (LSA)")
plt.show()
```

## (3) (10 points) Generate K topics by using  lda2vec, the number of topics K should be decided by the coherence score, then summarize what are the topics. You may refer the code here:

https://nbviewer.org/github/cemoody/lda2vec/blob/master/examples/twenty_newsgroups/lda2vec/lda2vec.ipynb

```python
pyLDAvis.enable_notebook()

top = 10
topic_to_topwords = {}

geetha_texts = [
    "Geetha is a data scientist who specializes in natural language processing and machine learning.",
    "In her free time, Geetha enjoys reading books on artificial intelligence and exploring new technologies.",
    "Geetha is dedicated to leveraging technology to solve real-world challenges and make a positive impact.",
    "She is currently engaged in a project involving the development of a chatbot for advanced NLP-based customer support."
    "Geetha has a strong background in computer science and has worked on various data analytics projects.",
    "In addition to her technical skills, Geetha is an avid problem solver and enjoys tackling complex challenges.",
    "She is known for her dedication and innovative approach to problem-solving.",
]

top = 10
topics = len(geetha_texts)

for k in range(topics):
    words = nltk.word_tokenize(geetha_texts[k])

    word_freq = nltk.FreqDist(words)
    top_words = [word for word, freq in word_freq.most_common(top)]

    msg = f'Topic {k} has top words: {", ".join(top_words)}'
    print(msg)
    topic_to_topwords[k] = top_words

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(" ".join(words))
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Topic {k} Word Cloud")
    plt.show()
```
